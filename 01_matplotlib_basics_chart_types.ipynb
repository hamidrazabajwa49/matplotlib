# Matplotlib Basics: Chart Types

Gallery of core Matplotlib chart types with minimal styling examples. Each section is self-contained and runnable independently.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Basic Line Plot

```python
x = [1, 2, 3, 4, 5, 6]
y = [5, 10, 15, 35, 25, 10]

plt.plot(x, y)
plt.show()
```

## 2. Line Plot with Labels and Title

```python
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
sales = [52, 31, 54, 23, 73]

plt.plot(days, sales)
plt.title('Sales Data')
plt.xlabel('Days of the Week')
plt.ylabel('Sales per Day')
plt.show()
```

## 3. Styled Line Plot (color, markers, grid, custom ticks)

```python
months = [1, 2, 3, 4, 5]
sales = [1000, 2000, 3000, 4000, 5000]

plt.plot(months, sales, color='blue', linestyle='--', linewidth=5,
         marker='o', label='Sales Report')
plt.xlabel('Months')
plt.ylabel('Sales per Month')
plt.title('Sales Report')
plt.legend(loc='upper left', fontsize=12)
plt.grid(color='gray', linestyle=':', linewidth=2)
plt.xticks([1, 2, 3, 4, 5], ['Mon1', 'Mon2', 'Mon3', 'Mon4', 'Mon5'])
plt.yticks([1000, 2000, 3000, 4000, 5000],
           [f'{v} Units' for v in [1000, 2000, 3000, 4000, 5000]])
plt.show()
```

## 4. Horizontal Bar Chart

```python
products = ['A', 'B', 'C', 'D', 'E']
sales = [1000, 2000, 3000, 4000, 5000]

plt.barh(products, sales, color='skyblue', label='Sales 2025')
plt.title('Sales Report')
plt.xlabel('Sales Units')
plt.ylabel('Products')
plt.legend()
plt.show()
```

## 5. Pie Chart

```python
products = ['A', 'B', 'C', 'D']
sales = [1000, 2000, 3000, 4000]

plt.pie(sales, labels=products, autopct='%1.1f%%',
        colors=['red', 'green', 'blue', 'gray'])
plt.title('Sales Distribution by Product')
plt.show()
```

## 6. Histogram

```python
scores = [42, 51, 85, 49, 33, 22, 27, 35, 37, 71, 77, 79, 83, 24, 29,
          18, 67, 59, 43, 50, 14, 15, 46, 56, 94, 25, 68, 84, 27, 19]

plt.hist(scores, bins=10, color='skyblue', edgecolor='gray')
plt.xlabel('Score Range')
plt.ylabel('Number of Students')
plt.title('Student Test Score Distribution')
plt.show()
```

## 7. Scatter Plot with Multiple Groups

```python
hours_studied = [5, 7, 9, 4, 6, 4, 3, 10]
scores = [42, 51, 85, 49, 33, 22, 27, 35]

plt.scatter(hours_studied, scores, color='skyblue', marker='^', label='Class 1')
plt.scatter([2, 4, 6, 8, 10, 12, 14], [1, 3, 5, 7, 9, 11, 13],
            color='red', marker='+', label='Class 2')
plt.legend()
plt.xlabel('Hours Studied')
plt.ylabel('Test Scores')
plt.title('Test Score Analysis')
plt.grid(color='gray', linestyle='--', linewidth=2)
plt.show()
```

## 8. Side-by-Side Subplots (plt.subplot)

```python
x = [2, 4, 6, 8, 7, 12, 14]
y = [1, 3, 5, 7, 3, 11, 13]

plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.title('Line Chart')

plt.subplot(1, 2, 2)
plt.bar(x, y)
plt.title('Bar Chart')

plt.tight_layout()
plt.show()
```

## 9. Side-by-Side Subplots (plt.subplots) with Figure Title

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

x = [2, 4, 6, 8, 7, 12, 14]
y = [1, 3, 5, 7, 3, 11, 13]

ax[0].plot(x, y)
ax[0].set_title('Line Plot')

ax[1].bar(x, y)
ax[1].set_title('Bar Chart')

fig.suptitle('Comparison of Charts')
plt.tight_layout()
plt.show()
```

## 10. Stacked Subplots: Sine vs Cosine

```python
x = np.linspace(0, 10, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

fig, axs = plt.subplots(2, 1)
axs[0].plot(x, y_sin, color='blue')
axs[0].set_title('Sine Wave')

axs[1].plot(x, y_cos, color='orange')
axs[1].set_title('Cosine Wave')

plt.tight_layout()
plt.show()
```
