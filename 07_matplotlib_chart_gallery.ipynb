# Matplotlib Chart Gallery

20 self-contained Matplotlib exercises covering common chart types: line, scatter, bar, histogram, pie, subplots, error bars, polar plots, and styling.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
```

## Problem 1: Basic Line Plot

```python
x = np.arange(1, 10, 0.5)
y_squared = x ** 2
y_cubed = x ** 3

plt.plot(x, y_squared, color='skyblue', marker='o', linestyle='--', label='X²')
plt.plot(x, y_cubed, color='orange', marker='s', linestyle='-', label='X³')
plt.xlabel('x')
plt.ylabel('Value')
plt.legend()
plt.grid(color='gray', linewidth=1.5)
plt.title('Problem 1: Line Plot')
plt.tight_layout()
plt.show()
```

## Problem 2: Scatter Plot

```python
arr1 = np.random.randint(1, 100, size=50)
arr2 = np.random.randint(1, 100, size=50)

plt.scatter(arr1, arr2, color='skyblue', marker='x', alpha=0.7)
plt.xlabel('Random Values 1')
plt.ylabel('Random Values 2')
plt.title('Problem 2: Scatter Plot')
plt.show()
```

## Problem 3: Bar Chart

```python
subjects = ['M1', 'M2', 'M3', 'M4', 'M5']
marks = [21, 15, 19, 24, 20]

plt.bar(subjects, marks, color='skyblue', width=0.4)
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.title('Problem 3: Bar Chart')
for i, v in enumerate(marks):
    plt.text(i, v + 0.5, str(v), ha='center')
plt.show()
```

## Problem 4: Histogram

```python
values = np.random.randint(1, 1000, size=1000)

plt.hist(values, bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Value Range')
plt.ylabel('Frequency')
plt.title('Problem 4: Histogram')
plt.show()
```

## Problem 5: Pie Chart

```python
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
colors = ['skyblue', 'orange', 'green', 'red']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Problem 5: Pie Chart')
plt.show()
```

## Problem 6: Multiple Subplots

A 2x2 grid combining line, scatter, bar, and histogram in one figure.

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

x = np.arange(1, 10, 0.5)
axs[0, 0].plot(x, x ** 2, color='skyblue')
axs[0, 0].set_title('Line Plot')

axs[0, 1].scatter(arr1, arr2, color='orange', alpha=0.7)
axs[0, 1].set_title('Scatter Plot')

axs[1, 0].bar(subjects, marks, color='green')
axs[1, 0].set_title('Bar Chart')

axs[1, 1].hist(values, bins=20, color='purple', edgecolor='black')
axs[1, 1].set_title('Histogram')

fig.suptitle('Problem 6: Multiple Subplots')
plt.tight_layout()
plt.show()
```

## Problem 7: Stacked Bar Chart

```python
series_a = [3, 5, 7, 9, 11]
series_b = [2, 4, 6, 8, 10]

plt.bar(subjects, series_a, label='A', color='skyblue')
plt.bar(subjects, series_b, bottom=series_a, label='B', color='orange')
plt.legend()
plt.title('Problem 7: Stacked Bar')
plt.show()
```

## Problem 8: Horizontal Bar Chart

```python
plt.barh(subjects, marks, color='purple')
plt.xlabel('Marks')
plt.ylabel('Subjects')
plt.title('Problem 8: Horizontal Bar')
plt.show()
```

## Problem 9: Area Plot

```python
x = np.arange(1, 6)
y = np.array([1, 4, 6, 8, 4])

plt.fill_between(x, y, color='skyblue', alpha=0.4)
plt.plot(x, y, color='black')
plt.title('Problem 9: Area Plot')
plt.show()
```

## Problem 10: Annotated Scatter Plot

```python
x = np.random.rand(10)
y = np.random.rand(10)
point_labels = [f'P{i}' for i in range(10)]

plt.scatter(x, y)
for i, txt in enumerate(point_labels):
    plt.annotate(txt, (x[i], y[i]))
plt.title('Problem 10: Annotated Scatter')
plt.show()
```

## Problem 11: Box Plot

```python
data = np.random.normal(0, 1, 100)

plt.boxplot(data)
plt.title('Problem 11: Box Plot')
plt.show()
```

## Problem 12: Violin Plot

```python
data = np.random.normal(0, 1, 100)

plt.violinplot(data)
plt.title('Problem 12: Violin Plot')
plt.show()
```

## Problem 13: Heatmap (imshow)

```python
data = np.random.rand(5, 5)

plt.imshow(data, cmap='viridis', interpolation='nearest')
plt.colorbar()
plt.title('Problem 13: Heatmap')
plt.show()
```

## Problem 14: Multiple Lines in One Plot

```python
x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')
plt.plot(x, np.sin(2 * x), label='sin(2x)')
plt.legend()
plt.title('Problem 14: Multiple Lines')
plt.show()
```

## Problem 15: Twin Axis Plot

```python
x = np.arange(1, 6)
y1 = x ** 2
y2 = x ** 3

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

ax1.plot(x, y1, 'g-')
ax2.plot(x, y2, 'b-')

ax1.set_xlabel('X')
ax1.set_ylabel('Square', color='g')
ax2.set_ylabel('Cube', color='b')

plt.title('Problem 15: Twin Axis')
plt.show()
```

## Problem 16: Step Plot

```python
x = np.arange(0, 5, 0.2)
y = np.exp(-x)

plt.step(x, y, where='mid', label='mid')
plt.legend()
plt.title('Problem 16: Step Plot')
plt.show()
```

## Problem 17: Error Bar Plot

```python
x = np.arange(0, 10, 1)
y = np.random.rand(10)
errors = np.random.rand(10) * 0.1

plt.errorbar(x, y, yerr=errors, fmt='o', capsize=5)
plt.title('Problem 17: Error Bars')
plt.show()
```

## Problem 18: Histogram with Gaussian Curve

```python
data = np.random.randn(1000)

count, bins, _ = plt.hist(data, bins=30, density=True, alpha=0.6, color='g')
plt.plot(bins, 1 / np.sqrt(2 * np.pi) * np.exp(-0.5 * bins ** 2),
         linewidth=2, color='black')
plt.title('Problem 18: Histogram with Gaussian')
plt.show()
```

## Problem 19: Polar Plot

```python
theta = np.linspace(0, 2 * np.pi, 100)
r = 1 + np.sin(6 * theta)

plt.polar(theta, r)
plt.title('Problem 19: Polar Plot')
plt.show()
```

## Problem 20: Custom Style Plot

```python
plt.style.use('ggplot')

x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')
plt.legend()
plt.title('Problem 20: Custom Style')
plt.show()
```
