# Synthetic Dataset Visualization

Generates a synthetic dataset (sales, regions, satisfaction, product categories) and demonstrates core Matplotlib plot types: line, scatter, bar, and histogram.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)
```

## Generate Synthetic Data

```python
dates = pd.date_range(start='2024-01-01', periods=100, freq='D')

sales = np.arange(100) * 1.5 + np.random.normal(loc=0, scale=20, size=100)

regions = np.random.choice(
    ['North', 'South', 'East', 'West'], size=100, p=[0.2, 0.3, 0.4, 0.1]
)

customer_satisfaction = np.random.uniform(low=65, high=100, size=100)

product_categories = np.random.choice(
    ['Electronics', 'Books', 'Home Goods', 'Apparel'], size=100
)

practice_df = pd.DataFrame({
    'date': dates,
    'sales': sales,
    'region': regions,
    'customer_satisfaction': customer_satisfaction,
    'product_category': product_categories,
})

practice_df.head()
```

```python
practice_df.info()
```

## Line Plot: Sales Over Time

```python
plt.figure(figsize=(10, 5))
plt.plot(practice_df['date'], practice_df['sales'], color='steelblue')
plt.title('Sales Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```

## Scatter Plot: Sales vs Customer Satisfaction

```python
plt.figure(figsize=(8, 5))
plt.scatter(practice_df['customer_satisfaction'], practice_df['sales'],
            color='coral', alpha=0.7)
plt.title('Sales vs Customer Satisfaction')
plt.xlabel('Customer Satisfaction')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```

## Bar Chart: Record Count per Region

```python
region_counts = practice_df['region'].value_counts()

plt.figure(figsize=(7, 5))
plt.bar(region_counts.index, region_counts.values, color='skyblue')
plt.title('Number of Records per Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

## Histogram: Sales Distribution

```python
plt.figure(figsize=(8, 5))
plt.hist(practice_df['sales'], bins=10, color='mediumseagreen', edgecolor='black')
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

## Average Sales by Region

```python
region_avg_sales = practice_df.groupby('region')['sales'].mean().sort_values(ascending=False)
region_avg_sales
```

## Average Customer Satisfaction by Product Category

```python
category_avg_satisfaction = (
    practice_df.groupby('product_category')['customer_satisfaction']
    .mean()
    .sort_values(ascending=False)
)
category_avg_satisfaction
```

```python
plt.figure(figsize=(7, 5))
category_avg_satisfaction.plot(kind='bar', color='orange')
plt.title('Average Customer Satisfaction by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Average Satisfaction')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```
